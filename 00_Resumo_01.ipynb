# Interface pyplot

## Introdução ao pyplot e suas funções

Matplotlib é uma biblioteca de plotagem em Python que produz figuras de qualidade em uma variedade de formatos impressos e ambientes interativos. `matplotlib.pyplot` é um conjunto de funções, e é geralmente importado sob o alias `plt`.

Cada função `pyplot` faz alguma alteração em uma figura, como criar uma figura, criar uma área de plotagem em uma figura, plotar algumas linhas em uma área de plotagem, decorar a plotagem com rótulos, etc.

Aqui estão algumas funções `pyplot` comumente usadas:

- `plt.figure()`: Cria uma nova figura.
- `plt.plot()`: Plota y versus x como linhas e/ou marcadores.
- `plt.scatter()`: Plota um gráfico de dispersão de y versus x.
- `plt.bar()`: Plota um gráfico de barras.
- `plt.xlabel()`: Define o rótulo do eixo x.
- `plt.ylabel()`: Define o rótulo do eixo y.
- `plt.title()`: Define um título para os eixos.
- `plt.grid()`: Configura as linhas da grade.
- `plt.legend()`: Coloca uma legenda no eixo.
- `plt.show()`: Exibe uma figura.


### Exemplos

```python
import matplotlib.pyplot as plt

# Base de dados:
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

# Plotando o gráfico de linha com o comando plt.plot
    # primiro argumento é referente ao eixo X
    # segundo argumento é referente ao eiyo Y
plt.plot(meses, vendas)

# Utilizando o comando xlabel & ylabel para atribuir título aos eixos X e Y respectivamente:
plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')

# Comando plt.title para atribuir título ao gráfico
plt.title('Crescimento das Vendas ao Longo do Ano')

# Comando show() para apresentar o gráfico na tela.
plt.show()
```

```python
# Mais alguns comandos:

    # Mais alguns atributos úteis para se utilizar no plt.plot:
        # marker: atribui um ponto nos encontros de X com Y 
            #(exemplos: 'o','d','x','s','v','+','p');
        # markersize: altera o tamanho do ponto (padrão é 6);
        # linestyle: altera o estilo da linha (padrão é '-') 
            # podendo ter como opção ('-','--','-.',':');
        # color: cor da linha
        # linewidth: "peso" da linha, grossura da linha.

plt.plot(meses, vendas, marker='o', markersize=5,  linestyle='-', color='green', linewidth=2)

    # Podemos realizar marcações no gráfico:
    # plt.annotate()
    # primeiro argumento é o Texto a ser exibido;
    # color: cor do texto;
    # xy: exibe um ponto onde o texto será exibido (localizando com os valores de X com Y);
    # xytext: exibe o ponto onde o texto será exibido;
    # arrowprops: adiciona uma seta APONTANDO para o 'xy'.
plt.annotate('Menor Valor', color='red', xy=(2, 78), xytext=(3, 80), arrowprops=dict(facecolor='red'))



# Para trazer linhas de grades: plt.grid(True | False) por padrão é False 
plt.grid(False)

plt.xlabel('Mês')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```
